```python
import os
import shutil
a = ['10_10', '15_5', '19_1']
c = ["train", "test"]
task2 = ["base", "new"]
root_folder = 'PASCAL_VOC'
for folder in a:
    for task in task2:
        for part in c:
            csv = pd.read_csv(f"csv_path/2007_{folder}_{part}_{task}.csv")
            for img, label in csv.values:
                old_dir = f"{root_folder}/images/{img}"
                new_dir = f"{folder}/{task}/images/{part}/{img}"
                shutil.copyfile(old_dir, new_dir)
                
                old_dir = f"{root_folder}/labels/{label}"
                new_dir = f"{folder}/{task}/labels/{part}/{label}"
                shutil.copyfile(old_dir, new_dir)
```
